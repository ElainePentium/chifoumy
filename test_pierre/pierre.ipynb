```python
import os
import cv2
import mediapipe as mp
import pandas as pd
```

```python
path = '../raw_data/paper/'
paper_image_files = []

for filename in os.listdir(path):
    f = os.path.join(path, filename)
    if os.path.isfile(f):
        paper_image_files.append(f)
```

```python
import matplotlib.pyplot as plt
```

```python
paper_image_files = ['../raw_data/paper/0a3UtNzl5Ll3sq8K.png', '../raw_data/paper/0cb6cVL8pkfi4wF6.png', '../raw_data/paper/0eqArS2GgsBeqgSn.png']
hands_list = []
mp_hands = mp.solutions.hands
with mp_hands.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.5) as hands:
    for idx, file in enumerate(paper_image_files):
        # Read an image, flip it around y-axis for correct handedness output (see above)
        image = cv2.flip(cv2.imread(file), 1)
        # plt.imshow(image)

        # Convert the BGR image to RGB before processing.
        img_process = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        # print(type(img_process))
        fingers = {}
        for i, finger in enumerate(img_process.multi_hand_landmarks[0].landmark, start=1):
            fingers[f'{i}x'] = (finger.x)
            fingers[f'{i}y'] = (finger.y)
        hands_list.append(fingers)

paper_df = pd.DataFrame(hands_list)
```
